### DataFrame.merge - объединение таблиц (аналог SQL: JOIN)

```python
# таблица с оценками
df_grades = pd.read_csv('grades.csv', index_col=0)
df_grades.head()
```

```python
# таблица хэшей: ФИО студента <-> хэш
df_hashes = pd.read_csv('hashes.csv')
df_hashes.head()
```

```python
df_grades.hash.nunique()
```

```python
df_hashes.hash.nunique()
```

```python
# присоединить подходящие строки из df_grades к df_hashes
df = pd.merge(df_hashes, df_grades, on='hash', how='left')
print(df.shape)
df.head(10)
```

```python
# присоединить подходящие строки из df_hashes к df_grades
df = pd.merge(df_hashes, df_grades, on='hash', how='right')
print(df.shape)
df.head(10)
```

```python
# пересечение таблиц
# в данном случае эквивалентно 'right', т.к. в df_grades нет таких хэшей, которые отсутствуют в df_hashes
df = pd.merge(df_hashes, df_grades, on='hash', how='inner')
print(df.shape)
df.head(10)
```

```python
# объединение таблиц
# в данном случае эквивалентно 'left', т.к. в df_grades нет таких хэшей, которые отсутствуют в df_hashes
df = pd.merge(df_hashes, df_grades, on='hash', how='outer')
print(df.shape)
df.head(10)
```

```python
# выбрать из таблицы хэшей только те строки, в которых хэш есть в таблице оценок
# т.е. отобрать тех студентов, которые писали контрольную и были оценены
df = df_hashes[df_hashes.hash.isin(df_grades.hash)]
print(df.shape)
df.head()
```

```python
# сколько человек из каждой группы были оценены?
df.Группа.value_counts()
```

### DataFrame.rolling - бегущее окно

```python
from datetime import datetime, timedelta
```

```python
df = pd.DataFrame(np.loadtxt('traj.txt'), columns=['x','y','z','vx','vy','vz','t'])
df.head()
```

```python
start_date = '2018-11-12 00:00:00'

def get_date(t, start):
    return datetime.fromisoformat(start) + timedelta(days=t*365/(2*np.pi))

df.index = df['t'].apply(get_date, args=(start_date,))
```

```python
df.head()
```

```python
xyz = df[['x','y','z']]
L = np.sqrt(np.sum((np.array(xyz[1:])-np.array(xyz[:-1]))**2, axis=1))*1.496000e+08
L = np.hstack(([0.], L))
df['L'] = L
```

```python
df.head()
```

```python
# расстояние, которое пролетел космический аппарат
plt.figure(figsize=(15,5))
df['L'][:'20260101'].plot()
```

```python
# усреднение на всех периодах по 180 дней
plt.figure(figsize=(15,5))
df.rolling('180d')['L'].mean().plot()
```

```python
# усреднение на последовательно увеличивающемся интервале (количестве строк)
plt.figure(figsize=(15,5))
df.expanding()['L'].mean().plot()
```

### [Reproducible Data Analysis in Jupyter](http://jakevdp.github.io/blog/2017/03/03/reproducible-data-analysis-in-jupyter/)

[Seattle Fremont Bridge Bike Counter](http://www.seattle.gov/transportation/projects-and-programs/programs/bike-program/bike-counters/fremont-bike-counters)

```python
import urllib3, json
from datetime import datetime, timedelta
```

```python
urls = ['http://www.eco-public.com/api/cw6Xk4jW4X4R/data/periode/101020981/?begin=20121008&end=20181112&step=3',
        'http://www.eco-public.com/api/cw6Xk4jW4X4R/data/periode/102020981/?begin=20121008&end=20181112&step=3']
```

```python
http = urllib3.PoolManager()
data = []
for url in urls:
    print(url, '...')
    response = http.request('GET', url)
    print('Status:', response.status)
    if response.status != 200:
        print('Info:', response.info())
    else:
        data.append(json.loads(response.data.decode(encoding='utf-8')))
```

```python
dfs = []
for d in data:
    dfs.append(pd.DataFrame(d))
```

```python
dfs[0].head()
```

```python
dfs[1].head()
```

```python
for df in dfs:
    df['d'] = df.timestamp.apply(lambda x:datetime.fromtimestamp(x/1000.0))
    df.drop(columns=['timestamp', 'date'], inplace=True)
```

```python
df = pd.merge(dfs[0], dfs[1], on='d', how='outer')
df.head()
```

```python
df.set_index('d', inplace=True)
df.head()
```

```python
df.columns = ['West', 'East']
df.head()
```

```python
df.resample('D').sum().rolling(365).sum().plot()#.set_ylim(0, None)
```

```python
df['Total'] = df.East + df.West
```

```python
df.groupby(df.index.time).mean().plot() #rolling('1H').sum()
```

```python
pivoted = df.pivot_table('Total', index=df.index.time, columns=df.index.date)
pivoted.iloc[:5,:5]
```

```python
pivoted.plot(legend=False, alpha=0.01, figsize=(15, 10))
```

```python
X = pivoted.fillna(0).T.values
X.shape
```

```python
from sklearn.decomposition import PCA
X2 = PCA(2, svd_solver='full').fit_transform(X)
X2.shape
```

```python
plt.scatter(X2[:,0], X2[:,1])
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(4)
gmm.fit(X)
labels = gmm.predict(X)
labels
```

```python
pivoted.shape, labels.shape
```

```python
s = pd.Series(labels)
```

```python
s.value_counts()
```

```python
plt.scatter(X2[:,0], X2[:,1], c=labels)
plt.colorbar()
```

```python
pivoted.T[labels==0].T.plot(legend=False, alpha=0.1)
```

```python
pivoted.T[labels==2].T.plot(legend=False, alpha=0.1)
```

```python
pivoted.shape
```

```python
dayofweek = pd.DatetimeIndex(pivoted.columns).dayofweek
dayofweek
```

```python
plt.scatter(X2[:,0], X2[:,1], c=dayofweek, cmap='rainbow')
plt.colorbar()
```
