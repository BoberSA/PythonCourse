## Что такое Python?

- в узком смысле - это интерпретируемый мультипарадигмальный язык программирования общего назначения
    - парадигмы:
        - функциональная,
        - процедурная,
        - объектно-ориентированная;


- в широком смысле - это целая экосистема, содержащая:
    - множество модулей, средств разработки, сервисов;
    - сообщество программистов, которые постоянно разрабатывают новые и улучшают имеющиеся модули,
    - сообщество пользователей, которые дают необходимую для разработки обратную связь.

```python
# Пример программы: загрузить все png файлы из текущей папки и вывести их на экран
import matplotlib.pyplot as plt
%matplotlib inline

img_list = [f for f in os.listdir() if f.endswith('.png')]

N, cols = len(img_list), 9
rows = (N + 1) // (cols + 1) + 1

plt.figure(figsize=(cols*2,rows*2))
plt.xlim(0, cols)
plt.ylim(0, rows)
plt.axis('off')

for i, png in enumerate(img_list):
    img = plt.imread(png)
    x, y, dy = i%cols, rows-1-i//cols, float(img.shape[0])/img.shape[1]
    plt.imshow(img, extent=[x, x+1, y, y+dy], interpolation='bilinear')
```
